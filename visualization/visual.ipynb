```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
from mpl_toolkits import mplot3d
```

```python
main_df = pd.read_csv("input00.txt", delimiter=' ', header=None)
x_df= main_df.iloc[2:,0]
y_df= main_df.iloc[2:,1]
fig = plt.figure()
ax = plt.axes()
ax.scatter(x_df, y_df)
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.title("The visual representation of the points")
plt.show()
```

```python
fig, ax = plt.subplots()
a1=[4,5,6]
b1=[5,7,4]
# Plot the data points
ax.scatter(a1, b1)

# Connect the points to form a closed polygon
ax.plot(a1 + [a1[0]], b1 + [b1[0]])

# Set the x and y axis limits
ax.set_xlim([min(a1) - 1, max(a1) + 1])
ax.set_ylim([min(b1) - 1, max(b1) + 1])

# Add grid lines
ax.grid(True)

# Show the plot
plt.show()
```
